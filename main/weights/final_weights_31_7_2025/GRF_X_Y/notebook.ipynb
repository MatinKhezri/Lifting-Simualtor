# Imports

```python
import numpy as np
import torch
import os
import torch.nn as nn
```

```python
class MyNN1(nn.Module):
  def __init__(self, input_dim, hidden1_dim, output_dim, activation_func):
    super(MyNN1, self).__init__()

    activation_functions = {
        'Tanh': nn.Tanh(),
        'ReLU': nn.ReLU(),
        'Sigmoid': nn.Sigmoid(),
        'ELU': nn.ELU(),
    }
    activation = activation_functions.get(activation_func)

    if activation is None:
        raise ValueError(f"Invalid activation function: {activation_func}")

    self.layers = nn.Sequential(
        nn.Linear(input_dim, hidden1_dim),
        nn.BatchNorm1d(hidden1_dim, momentum=0.1),
        activation,
        nn.Linear(hidden1_dim, output_dim)
    )

  def forward(self, x):
    return self.layers(x)
  



class MyNN2(nn.Module):
  def __init__(self, input_dim, hidden1_dim, hidden2_dim, output_dim, activation_func):
    super(MyNN2, self).__init__()

    activation_functions = {
    'Tanh': nn.Tanh(),
    'ReLU': nn.ReLU(),
    'Sigmoid': nn.Sigmoid(),
    'ELU': nn.ELU(),
    }
    activation = activation_functions.get(activation_func)

    if activation is None:
        raise ValueError(f"Invalid activation function: {activation_func}")


    self.layers = nn.Sequential(
        nn.Linear(input_dim, hidden1_dim),
        nn.BatchNorm1d(hidden1_dim),  # Added Batch Norm
        activation,
        nn.Linear(hidden1_dim, hidden2_dim),
        nn.BatchNorm1d(hidden2_dim),  # Added Batch Norm
        activation,
        nn.Linear(hidden2_dim, output_dim)
    )

  def forward(self, x):
    return self.layers(x)
  



class MyNN3(nn.Module):
  def __init__(self, input_dim, hidden1_dim, hidden2_dim, hidden3_dim, output_dim, activation_func):
    super(MyNN3, self).__init__()

    activation_functions = {
        'Tanh': nn.Tanh(),
        'ReLU': nn.ReLU(),
        'Sigmoid': nn.Sigmoid(),
        'ELU': nn.ELU(),
    }
    activation = activation_functions.get(activation_func)

    if activation is None:
        raise ValueError(f"Invalid activation function: {activation_func}")

    self.layers = nn.Sequential(
        nn.Linear(input_dim, hidden1_dim),
        nn.BatchNorm1d(hidden1_dim),  # Added Batch Norm
        activation,
        nn.Linear(hidden1_dim, hidden2_dim),
        nn.BatchNorm1d(hidden2_dim),  # Added Batch Norm
        activation,
        nn.Linear(hidden2_dim, hidden3_dim),
        nn.BatchNorm1d(hidden3_dim),  # Added Batch Norm
        activation,
        nn.Linear(hidden3_dim, output_dim)
    )

  def forward(self, x):
    return self.layers(x)
```
